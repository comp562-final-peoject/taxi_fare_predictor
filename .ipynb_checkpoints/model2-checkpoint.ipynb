```python
import pandas as pd
import numpy as np
import sklearn.metrics as skl
import matplotlib.pyplot as plt
import seaborn as sns
import math
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
%matplotlib inline
```

```python
data = pd.read_csv("data/cleaned_train_with_geo_distance.csv")
```

```python
features = ['year', 'hour', 'distance', 'passenger_count', 'isAirport']
input = data[features].values

#do the concatenate, so the first feature becomes the bias term
X = np.concatenate( ( np.ones((input.shape[0], 1)), input ), axis = 1 )

Y = data['fare_amount'].values
```

```python
l = data.shape[0]/4

data1 = X[l*0: l*(0+1)]
data2 = X[l*1: l*(1+1)]
data3 = X[l*2: l*(2+1)]
data4 = X[l*3: l*(3+1)]

train1_x = np.concatenate((data2, data3, data4), axis = 0)
train2_x = np.concatenate((data1, data3, data4), axis = 0)
train3_x = np.concatenate((data1, data2, data4), axis = 0)
train4_x = np.concatenate((data1, data2, data3), axis = 0)

train1_y = Y[l*1:]
train2_y = np.concatenate((Y[:l*1], Y[l*2:]),axis = 0)
train3_y = np.concatenate((Y[:l*2], Y[l*3:]),axis = 0)
train4_y = Y[:l*3]
```

```python
d = data.copy()
train = d.head(7000)
test = d.iloc[7000:]
```

```python
g = sns.FacetGrid(train, col="year", hue="passenger_count")
g.map(plt.scatter, "distance", "fare_amount")
g.add_legend()
```

```python
train['year'] = train['year']/2000
train
```

```python
train['hour'] = train['hour']/23
train
```

```python
X = train.loc[:, features].values
y = train.iloc[:, 0].values

X_test_dataset = test.loc[:, features].values

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
rand_forest_regressor = RandomForestRegressor()
rand_forest_regressor.fit(X_train, y_train)

y_rand_forest_predict = rand_forest_regressor.predict(X_test)
random_forest_model_error = math.sqrt(mean_squared_error(y_test, y_rand_forest_predict))
print(random_forest_model_error)
```

```python
XGB_model = XGBRegressor(learning_rate=0.3, max_depth=6, n_estimators=500)
XGB_model.fit(X_train, y_train)
y_XGB_predict = XGB_model.predict(X_test)

XGB_model_error = math.sqrt(mean_squared_error(y_test, y_XGB_predict))
print(XGB_model_error)
```
